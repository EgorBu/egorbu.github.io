```python
pylab.rcParams["figure.figsize"] = (12, 6)
data = zeros((900, 2))
random.seed(3)
data[:300, :2] = random.normal(size=(300, 2)) + [4, 4]
data[300:600, :2] = random.normal(size=(300, 2)) + [0, 0]
data[600:900, :2] = random.normal(size=(300, 2)) + [-4, 4]
xticks([])
yticks([])
scatter(data[:, 0], data[:, 1], c=[1] * len(data), cmap=cm.Greys)
c1 = Circle((0, 0), 2.6, fill=False, color="black")
c2 = Circle((4, 4), 2.6, fill=False, color="black")
c3 = Circle((-4, 4), 2.6, fill=False, color="black")
for c in (c1, c2, c3):
    gca().add_artist(c)
annotate("imposters", xy=(-4.5, 1.5), xytext=(-6.5, -1), size=24, arrowprops=dict(arrowstyle="->"))
annotate("juniors", xy=(0, 2.5), xytext=(-1, 5), size=24, arrowprops=dict(arrowstyle="->"))
annotate("awesome\nengineers", xy=(4.8, 1.6), xytext=(3.75, -1.5), size=24, arrowprops=dict(arrowstyle="->"))
savefig("pictures/motivation.png", bbox_inches="tight", dpi=100)
!pngquant -f pictures/motivation.png -o pictures/motivation.png
```
